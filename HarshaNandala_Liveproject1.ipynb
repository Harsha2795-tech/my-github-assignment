# Task 2: Data Cleaning

# Clean the dataset by standardizing column names, role names,
# CTC values, company type, education tier, work mode,
# removing duplicates, and handling missing values.

```python
df.columns = (
    df.columns
      .str.strip()
      .str.lower()
      .str.replace(" ", "_")
)
```

```python
sorted(df["role"].unique())
```

```python
role_mapping = {
    "Backend Engineer": "SDE Backend",
    "Backend Dev": "SDE Backend",
    "BE": "SDE Backend",
    "SDE Backend": "SDE Backend",

    "Frontend Engineer": "SDE Frontend",
    "Frontend Dev": "SDE Frontend",
    "SDE Frontend": "SDE Frontend",

    "Full Stack Engineer": "SDE Full-Stack",
    "Full Stack Developer": "SDE Full-Stack",
    "SDE Full Stack": "SDE Full-Stack"
}
```

```python
df["role_clean"] = df["role"].replace(role_mapping)
```

```python
df["role_clean"].value_counts()
```

```python
def parse_ctc(value):

    if pd.isna(value):
        return np.nan

    value = str(value).replace("₹", "").replace(",", "").replace("LPA", "").strip()

    try:
        number = float(value)

        if number > 100000:
            number = number / 100000

        return number

    except:
        return np.nan
```

```python
df["current_ctc_lpa"] = df["current_ctc"].apply(parse_ctc)

df["previous_ctc_lpa"] = df["previous_ctc"].apply(parse_ctc)
```

```python
df["company_type"].unique()
```

```python
company_mapping = {
    "UNICORN": "Unicorn",
    "unicorn": "Unicorn",
    "Unicorn": "Unicorn",

    "MNC": "MNC",
    "mnc": "MNC",

    "Mid-size": "Mid-size",
    "mid-size": "Mid-size",

    "Early-stage": "Early-stage",
    "early-stage": "Early-stage"
}

df["company_type_clean"] = df["company_type"].replace(company_mapping)
```

```python
df["company_type_clean"].value_counts()
```

```python
df["education_tier"].unique()
```

```python
education_mapping = {
    "Tier 1": "Tier 1",
    "Tier-1": "Tier 1",
    "T1": "Tier 1",
    "1": "Tier 1",

    "Tier 2": "Tier 2",
    "Tier-2": "Tier 2",
    "T2": "Tier 2",
    "2": "Tier 2",

    "Tier 3": "Tier 3",
    "Tier-3": "Tier 3",
    "T3": "Tier 3",
    "3": "Tier 3"
}

df["education_tier_clean"] = df["education_tier"].replace(education_mapping)
```

# Task 3: Business Analysis

## Q3.1: CTC Distribution by Role

# Calculate salary statistics for each role to identify
# which roles have the highest and lowest pay.

```python
role_salary = (
    df.groupby("role_clean")["current_ctc_lpa"]
      .agg(["median", "mean", "min", "max"])
      .sort_values(by="median", ascending=False)
)

print(role_salary)
```

### Insight (Q3.1)

Product Managers have the highest median salary of 31.60 LPA,
while Data Analysts have the lowest median salary of 15.90 LPA.

## Q3.2: Experience vs Salary

# Divide SDE Backend professionals into experience bands
# and calculate median salary for each band.

```python
backend = df[df["role_clean"] == "SDE Backend"].copy()

backend["exp_band"] = pd.cut(
    backend["years_exp"],
    bins=[-1, 1, 3, 5, 100],
    labels=["0-1", "2-3", "4-5", "6+"]
)

exp_salary = (
    backend.groupby("exp_band")["current_ctc_lpa"]
           .median()
)

print(exp_salary)
```

## Q3.3: Skill Premium
# Compare salaries of SDEs with and without selected skills.

```python
sde = df[df["role_clean"].isin([
    "SDE Backend",
    "SDE Frontend",
    "SDE Full-Stack"
])].copy()

skills = ["AWS", "ML", "System Design", "Kubernetes"]

for skill in skills:

    with_skill = sde[
        sde["skills"].str.contains(skill, case=False, na=False)
    ]["current_ctc_lpa"].median()

    without_skill = sde[
        ~sde["skills"].str.contains(skill, case=False, na=False)
    ]["current_ctc_lpa"].median()

    premium = ((with_skill - without_skill) / without_skill) * 100

    print(f"{skill}")
    print(f"With Skill : {with_skill:.2f}")
    print(f"Without Skill : {without_skill:.2f}")
    print(f"Premium : {premium:.2f}%")
    print("-"*40)
```

## Q3.4: Company Type Premium
# Compare salaries across different company types
# for SDE Backend professionals.

```python
backend = df[df["role_clean"] == "SDE Backend"]

company_salary = (
    backend.groupby("company_type_clean")["current_ctc_lpa"]
           .median()
)

print(company_salary)

unicorn = company_salary["Unicorn"]
mnc = company_salary["MNC"]

premium = ((unicorn - mnc) / mnc) * 100

print(f"\nUnicorn Premium over MNC = {premium:.2f}%")
```

## Q3.5: Underpaid Professionals

# Compare each employee's salary with the median salary
# of their role, company type, and experience band.

```python
df["exp_band"] = pd.cut(
    df["years_exp"],
    bins=[-1, 1, 3, 5, 100],
    labels=["0-1", "2-3", "4-5", "6+"]
)

group_sizes = (
    df.groupby(
        ["role_clean", "company_type_clean", "exp_band"]
    )["current_ctc_lpa"]
    .transform("count")
)

df_filtered = df[group_sizes >= 10].copy()

df_filtered["group_median"] = (
    df_filtered.groupby(
        ["role_clean", "company_type_clean", "exp_band"]
    )["current_ctc_lpa"]
    .transform("median")
)

df_filtered["gap"] = (
    df_filtered["current_ctc_lpa"] -
    df_filtered["group_median"]
)

underpaid = (
    df_filtered.sort_values("gap")
    [["employee_id",
      "role_clean",
      "company_type_clean",
      "years_exp",
      "current_ctc_lpa",
      "group_median",
      "gap"]]
    .head(10)
)

print(underpaid)
```
