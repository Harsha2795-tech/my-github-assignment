```python
# Step 1: Upload the WhatsApp chat file

from google.colab import files

uploaded = files.upload()
```

```python
# Get the uploaded file name
file_name = list(uploaded.keys())[0]

# Read the file
with open(file_name, "r", encoding="utf-8") as file:
    lines = file.readlines()

print("✅ File uploaded successfully!")
print("File Name:", file_name)
print("Total Lines:", len(lines))
```

```python
for i in range(10):
    print(lines[i].strip())
```

```python
# Step 2: Parse the chat

messages = []

system_count = 0
media_count = 0
deleted_count = 0

for line in lines:

    line = line.strip()

    if line == "":
        continue

    try:
        # Split into timestamp and remaining part
        timestamp, rest = line.split(" - ", 1)

        # If no sender exists, it is a system message
        if ": " not in rest:
            system_count += 1
            continue

        # Split sender and message
        sender, message = rest.split(": ", 1)

        # Split date and time
        date, time = timestamp.split(", ")

        # Count media messages
        if message == "<Media omitted>":
            media_count += 1

        # Count deleted messages
        if message == "This message was deleted":
            deleted_count += 1

        # Store every valid message
        messages.append({
            "date": date,
            "time": time,
            "sender": sender,
            "message": message
        })

    except:
        system_count += 1
```

```python
print("Total Parsed Messages :", len(messages))
print("System Messages       :", system_count)
print("Media Messages        :", media_count)
print("Deleted Messages      :", deleted_count)
```

```python
for msg in messages[:5]:
    print(msg)
```

```python
# Step 3: Group Overview

participants = {}
unique_people = set()

for msg in messages:
    sender = msg["sender"]

    unique_people.add(sender)

    if sender in participants:
        participants[sender] += 1
    else:
        participants[sender] = 1

# Sort by message count (highest first)
sorted_participants = sorted(participants.items(),
                             key=lambda x: x[1],
                             reverse=True)

# Date range
first_date = messages[0]["date"]
last_date = messages[-1]["date"]

print("="*55)
print("             GROUP OVERVIEW")
print("="*55)

print(f"Total Messages    : {len(messages)}")
print(f"Participants      : {len(unique_people)}")
print(f"First Message     : {first_date}")
print(f"Last Message      : {last_date}")

print("\nMessages Per Person")
print("-"*55)

for name, count in sorted_participants:
    percentage = (count / len(messages)) * 100
    print(f"{name:<10} : {count:>4} ({percentage:.2f}%)")
```

```python
# Step 4: Most Active Day and Hour

day_count = {}
hour_count = {}

for msg in messages:

    # Count messages per day
    date = msg["date"]

    if date in day_count:
        day_count[date] += 1
    else:
        day_count[date] = 1

    # Count messages per hour
    hour = int(msg["time"].split(":")[0])

    if hour in hour_count:
        hour_count[hour] += 1
    else:
        hour_count[hour] = 1


# Find busiest day
most_active_day = max(day_count, key=day_count.get)

# Find busiest hour
most_active_hour = max(hour_count, key=hour_count.get)


print("="*60)
print("          MOST ACTIVE DAY & HOUR")
print("="*60)

print(f"Most Active Day  : {most_active_day}")
print(f"Messages         : {day_count[most_active_day]}")

print()

print(f"Most Active Hour : {most_active_hour:02d}:00 - {most_active_hour:02d}:59")
print(f"Messages         : {hour_count[most_active_hour]}")
```

```python
# Step 5: Word Frequency Analysis

word_count = {}

stop_words = {
    "the", "is", "and", "to", "a", "an", "of", "in", "on",
    "for", "at", "it", "this", "that", "i", "you", "me",
    "my", "your", "we", "our", "hai", "bhai", "yaar",
    "kya", "ka", "ki", "ke", "ho", "haan", "na", "mat",
    "hai?", "ya", "are", "am", "was", "were"
}

for msg in messages:

    text = msg["message"]

    # Skip media and deleted messages
    if text == "<Media omitted>" or text == "This message was deleted":
        continue

    # Convert to lowercase
    text = text.lower()

    # Remove punctuation
    for ch in ".,!?;:()[]{}\"'<>/-":
        text = text.replace(ch, " ")

    # Split into words
    words = text.split()

    for word in words:

        if word in stop_words:
            continue

        if len(word) <= 2:
            continue

        if word in word_count:
            word_count[word] += 1
        else:
            word_count[word] = 1
```

```python
sorted_words = sorted(
    word_count.items(),
    key=lambda x: x[1],
    reverse=True
)
```

```python
print("="*55)
print("           TOP 20 MOST USED WORDS")
print("="*55)

for word, count in sorted_words[:20]:
    print(f"{word:<20} {count}")
```

```python
# Step 6: Message Statistics

person_stats = {}

for msg in messages:

    sender = msg["sender"]
    message = msg["message"]

    # Create dictionary for a new sender
    if sender not in person_stats:
        person_stats[sender] = {
            "messages": 0,
            "total_length": 0,
            "longest_message": "",
            "media": 0,
            "deleted": 0
        }

    # Total messages
    person_stats[sender]["messages"] += 1

    # Media messages
    if message == "<Media omitted>":
        person_stats[sender]["media"] += 1
        continue

    # Deleted messages
    if message == "This message was deleted":
        person_stats[sender]["deleted"] += 1
        continue

    # Message length
    length = len(message)
    person_stats[sender]["total_length"] += length

    # Longest message
    if length > len(person_stats[sender]["longest_message"]):
        person_stats[sender]["longest_message"] = message
```

```python
print("=" * 80)
print("MESSAGE STATISTICS")
print("=" * 80)

for person, stats in person_stats.items():

    average_length = (
        stats["total_length"] / stats["messages"]
    )

    print(f"\nPerson : {person}")
    print("-" * 50)
    print(f"Total Messages      : {stats['messages']}")
    print(f"Average Length      : {average_length:.2f} characters")
    print(f"Media Shared        : {stats['media']}")
    print(f"Deleted Messages    : {stats['deleted']}")
    print(f"Longest Message     :")
    print(stats["longest_message"])
```

```python
# Step 7: Behavior Analysis

behaviour = {}

for person in person_stats:

    behaviour[person] = {
        "night_messages": 0,
        "care_messages": 0,
        "drama_messages": 0
    }

# Words used for detection
care_words = [
    "take care", "eat", "sleep", "water",
    "notes", "help", "study", "home",
    "breakfast", "dinner"
]

drama_words = [
    "omg", "cry", "crying", "wow",
    "unbelievable", "help", "audacity",
    "shook", "angry", "screaming",
    "literally", "worst"
]

for msg in messages:

    sender = msg["sender"]
    message = msg["message"].lower()

    # Hour
    hour = int(msg["time"].split(":")[0])

    # Night messages (11 PM - 4 AM)
    if hour >= 23 or hour <= 4:
        behaviour[sender]["night_messages"] += 1

    # Caring messages
    for word in care_words:
        if word in message:
            behaviour[sender]["care_messages"] += 1

    # Drama messages
    for word in drama_words:
        if word in message:
            behaviour[sender]["drama_messages"] += 1
```

```python
# Spammer
spammer = max(person_stats,
              key=lambda x: person_stats[x]["messages"])

# Ghost
ghost = min(person_stats,
            key=lambda x: person_stats[x]["messages"])

# Night Owl
night_owl = max(
    behaviour,
    key=lambda x: behaviour[x]["night_messages"]
)

# Story Teller
story_teller = max(
    person_stats,
    key=lambda x:
    person_stats[x]["total_length"] /
    person_stats[x]["messages"]
)

# Group Mom
group_mom = max(
    behaviour,
    key=lambda x: behaviour[x]["care_messages"]
)

# Drama Queen
drama_queen = max(
    behaviour,
    key=lambda x: behaviour[x]["drama_messages"]
)
```

```python
print("="*60)
print("GROUP PERSONALITY REPORT")
print("="*60)

print(f" Spammer        : {spammer}")
print(f" Ghost          : {ghost}")
print(f" Night Owl      : {night_owl}")
print(f" Story Teller   : {story_teller}")
print(f" Group Mom      : {group_mom}")
print(f" Drama Queen    : {drama_queen}")
```

```python
print("\n" + "="*60)
print("PERSONALITY ARCHETYPES")
print("="*60)

for person in person_stats:

    if person == spammer:
        role = "📢 Spam Machine"

    elif person == ghost:
        role = "👻 Silent Observer"

    elif person == night_owl:
        role = "🌙 Night Owl"

    elif person == story_teller:
        role = "📚 Story Teller"

    elif person == group_mom:
        role = "❤️ Group Mom"

    elif person == drama_queen:
        role = "🎭 Drama Queen"

    else:
        role = "🙂 Active Member"

    print(f"{person:<10} ➜ {role}")
```

```python
print("\n")
print("=" * 70)
print("                GROUP DNA REPORT")
print("=" * 70)

# -------------------------------
# GROUP OVERVIEW
# -------------------------------

print("\n📌 GROUP OVERVIEW")
print("-" * 70)

print(f"Total Messages      : {len(messages)}")
print(f"Participants        : {len(unique_people)}")
print(f"First Message Date  : {first_date}")
print(f"Last Message Date   : {last_date}")
print(f"System Messages     : {system_count}")
print(f"Media Messages      : {media_count}")
print(f"Deleted Messages    : {deleted_count}")

# -------------------------------
# MESSAGE COUNT
# -------------------------------

print("\n📊 MESSAGE COUNT")
print("-" * 70)

for person, count in sorted_participants:
    percentage = (count / len(messages)) * 100
    print(f"{person:<10} : {count:>4} ({percentage:.2f}%)")

# -------------------------------
# ACTIVITY
# -------------------------------

print("\n📅 ACTIVITY")
print("-" * 70)

print(f"Most Active Day  : {most_active_day}")
print(f"Messages         : {day_count[most_active_day]}")

print()

print(f"Most Active Hour : {most_active_hour:02d}:00 - {most_active_hour:02d}:59")
print(f"Messages         : {hour_count[most_active_hour]}")

# -------------------------------
# TOP WORDS
# -------------------------------

print("\n📝 TOP 10 WORDS")
print("-" * 70)

for word, count in sorted_words[:10]:
    print(f"{word:<20} {count}")

# -------------------------------
# PERSONALITY
# -------------------------------

print("\n🎭 PERSONALITY REPORT")
print("-" * 70)

print(f"📢 Spammer        : {spammer}")
print(f"👻 Ghost          : {ghost}")
print(f"🌙 Night Owl      : {night_owl}")
print(f"📚 Story Teller   : {story_teller}")
print(f"❤️ Group Mom      : {group_mom}")
print(f"🎭 Drama Queen    : {drama_queen}")

# -------------------------------
# INDIVIDUAL STATS
# -------------------------------

print("\n👥 INDIVIDUAL STATISTICS")
print("-" * 70)

for person, stats in person_stats.items():

    average = stats["total_length"] / stats["messages"]

    print(f"\n{person}")
    print(f"Messages          : {stats['messages']}")
    print(f"Average Length    : {average:.2f}")
    print(f"Media Shared      : {stats['media']}")
    print(f"Deleted Messages  : {stats['deleted']}")

print("\n")
print("=" * 70)
print("            END OF GROUP DNA REPORT")
print("=" * 70)
```
